```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import json
from time import sleep

url = "https://www.apartmentguide.com/apartments/Illinois/Chicago/"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/139.0.7258.128 Safari/537.36",
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8"
}

r = requests.get(url, headers=headers)
soup = BeautifulSoup(r.text, 'lxml')

# находим все карточки на странице
cards = soup.find_all('li', class_='_90363ab6')

data = []

for card in cards:
    try:
        price = card.find('div', class_='_3dc1722d').find('div', class_='_a3684b85').find('span').text.strip()
    except:
        price = None

    try:
        link = 'https://www.apartmentguide.com' + card.find('a')['href']
    except:
        link = None

    if link:
        r_2 = requests.get(link, headers=headers)
        sleep(2)  # чтобы не забанили
        soup_2 = BeautifulSoup(r_2.text, 'lxml')

        # beds, baths, sqft
        try:
            bads = soup_2.find('li', {'data-tid': 'pdpKeyInfo_bedText'}).text.strip()
        except:
            bads = None

        try:
            baths = soup_2.find('li', {'data-tid': 'pdpKeyInfo_bathText'}).text.strip()
        except:
            baths = None

        try:
            sqft = soup_2.find('li', {'data-tid': 'pdpKeyInfo_sqFt'}).text.strip()
        except:
            sqft = None

        # pets
        try:
            pets = [s.get_text(strip=True)
                    for s in soup_2.find('div', attrs={'data-tag_section': 'pet_policy_summary'}) \
                                 .find_all('span')]
            pets = ', '.join(pets) if pets else None
        except:
            pets = None

        # amenities
        sections = ['Kitchen', 'Features', 'Unique', 'Parking']
        amenities_data = {}
        for section in sections:
            try:
                block = soup_2.find('h3', text=section).find_next('ul')
                items = [li.text.strip() for li in block.find_all('li')]
                amenities_data[section] = ', '.join(items) if items else None
            except:
                amenities_data[section] = None

        # term length
        try:
            term_length = soup_2.find('h3', string='Term Length').find_next('div').text.strip()
        except:
            term_length = None

        # walkable / bikeable / transit
        def get_score(soup, keyword):
            try:
                return int(soup.find('h3', string=lambda x: x and keyword in x)
                               .find_next('span', class_='_047c9821')
                               .text.strip())
            except:
                return None

        walkable = get_score(soup_2, 'Walkable')
        bikeable = get_score(soup_2, 'Bikeable')
        transit = get_score(soup_2, 'Transit')

        # coordinates + postal_code
        try:
            script_tag = soup_2.find('script', {'id': 'json-ld', 'type': 'application/ld+json'})
            json_data = json.loads(script_tag.string)

            latitude = json_data["mainEntity"]["geo"]["latitude"]
            longitude = json_data["mainEntity"]["geo"]["longitude"]
            postal_code = json_data['mainEntity']['address']['postalCode']
        except:
            latitude = longitude = postal_code = None

        # собираем данные
        data.append([
            price, bads, baths, sqft, pets,
            amenities_data['Kitchen'], amenities_data['Features'],
            amenities_data['Unique'], amenities_data['Parking'],
            term_length, walkable, bikeable, transit,
            latitude, longitude, postal_code, link
        ])
```

```python
# создаем DataFrame
columns = [
    'price', 'bads', 'baths', 'sqft', 'pets',
    'kitchen', 'features', 'unique', 'parking',
    'term_length', 'walkable', 'bikeable', 'transit',
    'latitude', 'longitude', 'postal_code', 'link'
]

df = pd.DataFrame(data, columns=columns)
df.head()
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import json
from time import sleep

base_url = "https://www.apartmentguide.com/apartments/Illinois/Chicago/page-{}"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/139.0.7258.128 Safari/537.36",
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8"
}

data = []
page = 1

while True:
    url = base_url.format(page) if page > 1 else "https://www.apartmentguide.com/apartments/Illinois/Chicago/"
    print(f"Парсим страницу {page}: {url}")
    
    r = requests.get(url, headers=headers)
    soup = BeautifulSoup(r.text, 'lxml')
    
    cards = soup.find_all('li', class_='_90363ab6')
    if not cards:
        print("Карточки закончились, выходим.")
        break

    for card in cards:
        try:
            price = card.find('div', class_='_3dc1722d').find('div', class_='_a3684b85').find('span').text.strip()
        except:
            price = None

        try:
            link = 'https://www.apartmentguide.com' + card.find('a')['href']
        except:
            link = None

        if link:
            r_2 = requests.get(link, headers=headers)
            sleep(2)
            soup_2 = BeautifulSoup(r_2.text, 'lxml')

            # beds, baths, sqft
            try: bads = soup_2.find('li', {'data-tid': 'pdpKeyInfo_bedText'}).text.strip()
            except: bads = None
            try: baths = soup_2.find('li', {'data-tid': 'pdpKeyInfo_bathText'}).text.strip()
            except: baths = None
            try: sqft = soup_2.find('li', {'data-tid': 'pdpKeyInfo_sqFt'}).text.strip()
            except: sqft = None

            # pets
            try:
                pets = [s.get_text(strip=True)
                        for s in soup_2.find('div', attrs={'data-tag_section': 'pet_policy_summary'}).find_all('span')]
                pets = ', '.join(pets) if pets else None
            except: pets = None

            # amenities
            sections = ['Kitchen', 'Features', 'Unique', 'Parking']
            amenities_data = {}
            for section in sections:
                try:
                    block = soup_2.find('h3', text=section).find_next('ul')
                    items = [li.text.strip() for li in block.find_all('li')]
                    amenities_data[section] = ', '.join(items) if items else None
                except: amenities_data[section] = None

            # term length
            try: term_length = soup_2.find('h3', string='Term Length').find_next('div').text.strip()
            except: term_length = None

            # walkable / bikeable / transit
            def get_score(soup, keyword):
                try:
                    return int(soup.find('h3', string=lambda x: x and keyword in x)
                               .find_next('span', class_='_047c9821')
                               .text.strip())
                except: return None

            walkable = get_score(soup_2, 'Walkable')
            bikeable = get_score(soup_2, 'Bikeable')
            transit = get_score(soup_2, 'Transit')

            # coordinates + postal_code
            try:
                script_tag = soup_2.find('script', {'id': 'json-ld', 'type': 'application/ld+json'})
                json_data = json.loads(script_tag.string)
                latitude = json_data["mainEntity"]["geo"]["latitude"]
                longitude = json_data["mainEntity"]["geo"]["longitude"]
                postal_code = json_data['mainEntity']['address']['postalCode']
            except:
                latitude = longitude = postal_code = None

            # собираем данные
            data.append([
                price, bads, baths, sqft, pets,
                amenities_data['Kitchen'], amenities_data['Features'],
                amenities_data['Unique'], amenities_data['Parking'],
                term_length, walkable, bikeable, transit,
                latitude, longitude, postal_code, link
            ])
    
    page += 1 
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import time
import json

# ---------- Настройки Selenium ----------
chrome_options = Options()
chrome_options.add_argument("--window-size=1920,1080")
# Для теста используем обычный режим, не headless, чтобы гарантированно загрузились все элементы
# chrome_options.add_argument("--headless")  # можно включить позже

service = Service("/Applications/chromedriver-mac-x64/chromedriver")  # путь к chromedriver
driver = webdriver.Chrome(service=service, options=chrome_options)

base_url = "https://www.apartmentguide.com/apartments/Illinois/Chicago/page-{}/"
data = []
page = 1

while True:
    url = base_url.format(page) if page > 1 else "https://www.apartmentguide.com/apartments/Illinois/Chicago/"
    print(f"Парсим страницу {page}: {url}")
    driver.get(url)
    
    # Ждём появления карточек
    try:
        WebDriverWait(driver, 10).until(
            EC.presence_of_all_elements_located((By.CSS_SELECTOR, "li[data-tid^='srp_card_']"))
        )
    except:
        print("Карточки не найдены, выходим.")
        break
    
    # Скроллим вниз, чтобы подгрузились все карточки
    last_height = driver.execute_script("return document.body.scrollHeight")
    while True:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(2)
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height
    
    cards = driver.find_elements(By.CSS_SELECTOR, "li[data-tid^='srp_card_']")
    if not cards:
        print("Карточки закончились, выходим.")
        break

    print(f"Найдено карточек на странице: {len(cards)}")
    
    for card in cards:
        try:
            price = card.find_element(By.CSS_SELECTOR, "div[data-name='PriceText'] span").text
        except:
            price = None
        try:
            link = card.find_element(By.TAG_NAME, "a").get_attribute("href")
        except:
            link = None

        if link:
            driver.get(link)
            try:
                WebDriverWait(driver, 10).until(
                    EC.presence_of_element_located((By.CSS_SELECTOR, "li[data-tid^='pdpKeyInfo']"))
                )
            except:
                continue
            
            # beds, baths, sqft
            try: bads = driver.find_element(By.CSS_SELECTOR, "li[data-tid='pdpKeyInfo_bedText']").text
            except: bads = None
            try: baths = driver.find_element(By.CSS_SELECTOR, "li[data-tid='pdpKeyInfo_bathText']").text
            except: baths = None
            try: sqft = driver.find_element(By.CSS_SELECTOR, "li[data-tid='pdpKeyInfo_sqFt']").text
            except: sqft = None
            
            # pets
            try:
                pets_elems = driver.find_elements(By.CSS_SELECTOR, "div[data-tag_section='pet_policy_summary'] span")
                pets = ', '.join([e.text for e in pets_elems]) if pets_elems else None
            except: pets = None
            
            # amenities
            sections = ['Kitchen', 'Features', 'Unique', 'Parking']
            amenities_data = {}
            for section in sections:
                try:
                    heading = driver.find_element(By.XPATH, f"//h3[text()='{section}']")
                    ul = heading.find_element(By.XPATH, "following-sibling::ul")
                    items = [li.text for li in ul.find_elements(By.TAG_NAME, "li")]
                    amenities_data[section] = ', '.join(items) if items else None
                except:
                    amenities_data[section] = None
            
            # term length
            try:
                term_length = driver.find_element(By.XPATH, "//h3[text()='Term Length']/following-sibling::div").text
            except:
                term_length = None
            
            # walkable / bikeable / transit
            def get_score(keyword):
                try:
                    return int(driver.find_element(By.XPATH,
                        f"//h3[contains(text(), '{keyword}')]/following-sibling::span").text)
                except:
                    return None
            walkable = get_score("Walkable")
            bikeable = get_score("Bikeable")
            transit = get_score("Transit")
            
            # coordinates + postal_code
            try:
                script_tag = driver.find_element(By.CSS_SELECTOR, "script#json-ld[type='application/ld+json']").get_attribute("innerHTML")
                json_data = json.loads(script_tag)
                latitude = json_data["mainEntity"]["geo"]["latitude"]
                longitude = json_data["mainEntity"]["geo"]["longitude"]
                postal_code = json_data['mainEntity']['address']['postalCode']
            except:
                latitude = longitude = postal_code = None
            
            data.append([
                price, bads, baths, sqft, pets,
                amenities_data['Kitchen'], amenities_data['Features'],
                amenities_data['Unique'], amenities_data['Parking'],
                term_length, walkable, bikeable, transit,
                latitude, longitude, postal_code, link
            ])
    
    page += 1
    # Для теста ограничимся 20 страницами (можно убрать)
    if page > 20:
        break

driver.quit()

# ---------- Сохраняем результат ----------
df = pd.DataFrame(data, columns=[
    'price','beds','baths','sqft','pets',
    'Kitchen','Features','Unique','Parking',
    'term_length','walkable','bikeable','transit',
    'latitude','longitude','postal_code','link'
])
df.to_csv('apartments_chicago.csv', index=False)
print("Сбор данных завершён. Всего квартир:", len(df))
```

```python
import undetected_chromedriver as uc
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import time
import json
import random

# ---------- Настройки undetected-chromedriver ----------
options = uc.ChromeOptions()
options.add_argument("--window-size=1920,1080")
# options.add_argument("--headless")  # можно включить позже
options.add_argument("--disable-blink-features=AutomationControlled")

driver = uc.Chrome(options=options)

# ---------- Настройки ----------
base_url = "https://www.apartmentguide.com/apartments/Illinois/Chicago/page-{}/"
data = []
page = 1

while True:
    url = base_url.format(page) if page > 1 else "https://www.apartmentguide.com/apartments/Illinois/Chicago/"
    print(f"Парсим страницу {page}: {url}")
    driver.get(url)
    
    # ---------- Скроллим страницу ----------
    last_height = driver.execute_script("return document.body.scrollHeight")
    for y in range(0, last_height, random.randint(100, 300)):
        driver.execute_script(f"window.scrollTo(0, {y});")
        time.sleep(random.uniform(0.2, 0.5))
    
    # ---------- Ждём карточки ----------
    try:
        WebDriverWait(driver, 10).until(
            EC.presence_of_all_elements_located((By.CSS_SELECTOR, "li[data-tid^='srp_card_']"))
        )
    except:
        print("Карточки не найдены, выходим со страницы.")
        break
    
    cards = driver.find_elements(By.CSS_SELECTOR, "li[data-tid^='srp_card_']")
    if not cards:
        print("Карточки закончились, выходим.")
        break
    
    print(f"Найдено карточек на странице: {len(cards)}")
    
    # ---------- Сначала собираем ссылки ----------
    card_links = []
    for card in cards:
        try:
            link = card.find_element(By.TAG_NAME, "a").get_attribute("href")
            if link:
                card_links.append(link)
        except:
            continue
    print(f"Собрано ссылок на карточки: {len(card_links)}")
    
    # ---------- Парсим каждую карточку ----------
    for link in card_links:
        driver.get(link)
        time.sleep(random.uniform(1, 2))  # чтобы не заблокировали
        try:
            WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.CSS_SELECTOR, "li[data-tid^='pdpKeyInfo']"))
            )
        except:
            continue
        
        # ---------- Price ----------
        try:
            price = driver.find_element(By.CSS_SELECTOR, "div._51bb0611 div[data-tid='pdpKeyInfo_price']").text.strip()
        except:
            price = None
        
        # ---------- beds, baths, sqft ----------
        try: bads = driver.find_element(By.CSS_SELECTOR, "li[data-tid='pdpKeyInfo_bedText']").text
        except: bads = None
        try: baths = driver.find_element(By.CSS_SELECTOR, "li[data-tid='pdpKeyInfo_bathText']").text
        except: baths = None
        try: sqft = driver.find_element(By.CSS_SELECTOR, "li[data-tid='pdpKeyInfo_sqFt']").text
        except: sqft = None
        
        # ---------- pets ----------
        try:
            pets_elems = driver.find_elements(By.CSS_SELECTOR, "div[data-tag_section='pet_policy_summary'] span")
            pets = ', '.join([e.text for e in pets_elems]) if pets_elems else None
        except: pets = None
        
        # ---------- amenities ----------
        sections = ['Kitchen', 'Features', 'Unique', 'Parking']
        amenities_data = {}
        for section in sections:
            try:
                heading = driver.find_element(By.XPATH, f"//h3[text()='{section}']")
                ul = heading.find_element(By.XPATH, "following-sibling::ul")
                items = [li.text for li in ul.find_elements(By.TAG_NAME, "li")]
                amenities_data[section] = ', '.join(items) if items else None
            except:
                amenities_data[section] = None
        
        # ---------- term length ----------
        try:
            term_length = driver.find_element(By.XPATH, "//h3[text()='Term Length']/following-sibling::div").text
        except:
            term_length = None
        
        # ---------- coordinates + postal_code ----------
        try:
            script_tag = driver.find_element(By.CSS_SELECTOR, "script#json-ld[type='application/ld+json']").get_attribute("innerHTML")
            json_data = json.loads(script_tag)
            latitude = json_data["mainEntity"]["geo"]["latitude"]
            longitude = json_data["mainEntity"]["geo"]["longitude"]
            postal_code = json_data['mainEntity']['address']['postalCode']
        except:
            latitude = longitude = postal_code = None
        
        data.append([
            price, bads, baths, sqft, pets,
            amenities_data['Kitchen'], amenities_data['Features'],
            amenities_data['Unique'], amenities_data['Parking'],
            term_length, latitude, longitude, postal_code, link
        ])
        print(f"✓ Спарсил: {link} | Price: {price}")
    
    # ---------- Пауза перед следующей страницей ----------
    time.sleep(random.uniform(3, 5))
    page += 1

driver.quit()

# ---------- Сохраняем результат ----------
df = pd.DataFrame(data, columns=[
    'price','beds','baths','sqft','pets',
    'Kitchen','Features','Unique','Parking',
    'term_length','latitude','longitude','postal_code','link'
])
df.to_csv('apartments_chicago_stealth_with_prices.csv', index=False, encoding="utf-8-sig")
print("Сбор данных завершён. Всего квартир:", len(df))
```
